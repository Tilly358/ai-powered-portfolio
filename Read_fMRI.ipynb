```python
# Specific prompt to ask for the scan interpretation.

# Ask for explanations on how it "reads" the scan.

prompt_text = """
You are an AI assistant specialized in medical imaging. I have uploaded an image of a brain scan.
Please interpret the key features of this brain scan.
Specifically, tell me:

1. What type of scan is this?
2. What planes of view are shown?
3. Point out and describe any prominent anatomical structures visible.
4. Are there any obvious abnormalities? If so, describe them without making a diagnosis.
5. Crucially, explain *how you are reading* the image to arrive at these observations.
6. What is the purpose of this scan?
7. How the machine function to get this scan?
8. As we get the scan result, what should we do first before starting reading the scan?
9. What should humans do after getting the analysis of the scan?
10. How should humans verify the reliability of the analysis?

Remember, this is for educational purposes, not medical diagnosis.
"""


response = model.generate_content([prompt_text, img])

print(response.text)
```
